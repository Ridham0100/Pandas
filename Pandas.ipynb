```python
terror = pd.read_csv(r"C:\Users\Hp\Downloads\terrorismData.csv")
terror.tail()
```

```python
df = terror.copy()
```

```python
mask = df['State'] == 'Jammu and Kashmir'
```

```python
filtered_data = df[mask]
filtered_data
```

```python
grp_mask = filtered_data['Group'] != 'Unknown'
```

```python
filtered_data = filtered_data[grp_mask]
```

```python
filtered_data
```

```python
city_counts = filtered_data['City'].value_counts()
city_counts
```

```python
most_attacked_city = city_counts.idxmax()
most_attacked_city
```

```python
attacks_count = city_counts.max()
attacks_count
```

```python
grp_mask = filtered_data.loc[df['City'] == most_attacked_city, 'Group']
grp_mask
```

```python
active_grp = grp_mask.value_counts().idxmax()
active_grp
```

```python
print(most_attacked_city, attacks_count, active_grp , sep = ' ')
```
